## Exploring Antarctic Ozone Data
Let's import some data for the size of the ozone hole over Antarctica and minimum yearly ozone readings over time. Then let's make some plots to explore this data. The data is from https://ozonewatch.gsfc.nasa.gov/statistics/ytd_data.txt and was sent to me by Dr. Steven Payne.

```python
# import some packages/modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('https://raw.githubusercontent.com/prof-groff/evns-462/master/data/antarctic_ozone.csv')
dataframe.head() # dataframe  print(dataframe)
```

### Let's plot the ozone data as a function of time.

```python
plt.figure(figsize=(8,6))
plt.tight_layout()
plt.plot(dataframe['YEAR'], dataframe['MIN OZONE VALUE (DU)'], color='m', linewidth=2) # plt.scatter
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('year',fontsize=14)
plt.ylabel('minimum ozone (DU)', fontsize=14)
plt.savefig('figname.png',dpi=400)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.tight_layout()
plt.plot(dataframe['YEAR'], dataframe['OZONE HOLE AREA (MIL SQ. KM)'], color='g', linewidth=2) # plt.scatter
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('year',fontsize=14)
plt.ylabel('ozone hole area (Mkm$^2$)', fontsize=14)
plt.savefig('figname.png',dpi=400)
plt.show()
```

### Let's plot both datasets on the same figure suing subplots.

```python
# side by side
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,6))
fig.tight_layout()
ax1.plot(dataframe['YEAR'], dataframe['MIN OZONE VALUE (DU)'], color='m')
ax1.set_xlabel('year')
ax1.set_ylabel('minimum ozone (DU)')
ax2.plot(dataframe['YEAR'], dataframe['OZONE HOLE AREA (MIL SQ. KM)'], color='g')
ax2.set_xlabel('year')
ax2.set_ylabel('ozone hole area (Mkm$^2$)')
plt.show()
```

```python
# stacked
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,6))
fig.tight_layout()
ax1.plot(dataframe['YEAR'], dataframe['MIN OZONE VALUE (DU)'], color='m')
ax1.set_xlabel('year')
ax1.set_ylabel('minimum ozone (DU)')
ax2.plot(dataframe['YEAR'], dataframe['OZONE HOLE AREA (MIL SQ. KM)'], color='g')
ax2.set_xlabel('year')
ax2.set_ylabel('ozone hole area (Mkm$^2$)')
plt.show()
```

### Let's show both datasets on the same plot using different y-axis scales.

```python
fig, ax1 = plt.subplots(figsize=(8,6))
ax1.plot(dataframe['YEAR'], dataframe['MIN OZONE VALUE (DU)'], color='m')
ax1.set_ylabel('minimum ozone (DU)',color='m')
ax1.tick_params(axis='y', labelcolor='m')
ax1.set_xlabel('year')

ax2 = ax1.twinx()
ax2.plot(dataframe['YEAR'], dataframe['OZONE HOLE AREA (MIL SQ. KM)'],color='g')
ax2.set_ylabel('ozone hole area (Mkm$^2$)',color='g')
ax2.tick_params(axis='y', labelcolor='g')

ax1.legend(labels=['minimum ozone'])
ax2.legend(labels=['ozone hole area'],loc='upper left')
plt.title('Antarctic Ozone Over Time')

plt.show()

```
